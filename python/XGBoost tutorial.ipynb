```python
# numpy
import numpy as np
# data frames
import pandas as pd
# XGBoost
import xgboost as xgb
# Plotting
import matplotlib.pyplot as plt
%matplotlib inline
```

## Example data set

Let's use diabetes from Skilearn: breast cancer diagnosis based on tumor properties from images.

```python
from sklearn import datasets
```

```python
ds = datasets.load_breast_cancer(as_frame=True)
```

```python
print(ds.DESCR)
```

```python
ds.frame.head()
# ds.data.head()
# ds.target.head()
```

```python
ds.frame['target'].value_counts()
```

```python
ds.frame.info()
```

```python
ds.frame.describe()
```

```python
X, y = ds.frame.iloc[:,:-1], ds.frame.iloc[:,-1]
# X = ds.data
# y = ds.target
```

## Train-test split

Make an 80%-20% split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=6734)
```

## Gradient boosting decision tree
Define and tran the classifier

```python
xgb.XGBClassifier?
```

```python
xgb_classifier = xgb.XGBClassifier(n_estimators=10, max_depth=5, learning_rate=0.1)
```

train the `xgb_classifier` instance

```python
xgb_classifier.fit(X_train, y_train)
```

```python
y_pred = xgb_classifier.predict(X_test)
```

```python
y_pred
```

```python
np.array(y_test)
```

No skill classifier

```python
ns_pred = [0 for _ in range(len(y_test))]
```

```python
from sklearn.metrics import roc_auc_score
```

```python
ns_auc = roc_auc_score(y_test, ns_pred)
cl_auc = roc_auc_score(y_test, y_pred)
```

```python
print('No Skill: ROC AUC=%.3f' % (ns_auc))
print('Classifier: ROC AUC=%.3f' % (cl_auc))
```

## Plotting

```python
xgb.plot_tree(xgb_classifier, num_trees=0, fontsize=10)
fig = plt.gcf()
fig.set_size_inches(20, 10)
plt.show()
```

```python
xgb.plot_importance(xgb_classifier)
fig = plt.gcf()
fig.set_size_inches(20, 10)
plt.show()
```

```python
from sklearn.metrics import roc_curve
```

```python
ns_fpr, ns_tpr, _ = roc_curve(y_test, ns_pred)
cl_fpr, cl_tpr, _ = roc_curve(y_test, y_pred)
```

plot the roc curve for the model

```python
plt.rcParams['figure.figsize'] = [7, 5]
plt.plot(ns_fpr, ns_tpr, linestyle='--', label='No Skill')
plt.plot(cl_fpr, cl_tpr, marker='.', label='Logistic')
plt.title('ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```
